```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
    
import os
import pickle
```

```python
bok = pd.read_excel("BokarizaHydroSummary_WY12-22.xlsx")
```

```python
bok
```

```python
x=bok["Precipitation (inches per year)"]
y=bok["Infiltration (AFY)"]
n=bok["Water year"]

#Plot
plt.figure(figsize=(5, 6))
plt.scatter(x, y, marker='o', s=2)
plt.title("Runoff Model of ReNeM's BD Project ", fontsize="x-large")
plt.xlabel("Precipitation (inches per year)", fontsize="large")
plt.ylabel("Infiltration (AFY)", fontsize="large")
plt.xlim([0,40])
plt.ylim([0,170])
plt.errorbar(x, y, yerr=.15*y, fmt="o", linewidth=1, color = 'black')
plt.errorbar(x, y, xerr=.05*x, fmt="o", linewidth=1, color = 'black')
for i, txt in enumerate(n):
    plt.annotate(txt, (x[i], y[i]), rotation=0)
plt.axvline(x=21.6, color = '#1f77b4', linestyle="--", linewidth=1)
plt.text(20, 5, r'Historical', fontsize="large", color = '#1f77b4', rotation=90)
plt.text(22, 5, r'Average', fontsize="large", color = '#1f77b4', rotation=90)
plt.axhline(y=100, color = '#ff7f0e', linestyle="--", linewidth=1)
plt.text(1, 102, r'Project Goal', fontsize="large", color = '#ff7f0e')
plt.text(1, 93, r'100 AFY', fontsize="large", color = '#ff7f0e')

sns.regplot(x, y, order=1, color = 'gray', truncate=False)

plt.savefig("Bokariza.png") 
```
